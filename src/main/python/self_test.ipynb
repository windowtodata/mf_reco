```python
import configparser
import mlflow
import json
import os
mlflow.__version__
```

```python
config = configparser.ConfigParser()
config.read("../resources/train_params.ini")
```

```python
mlflow.set_tracking_uri(config.get("mlflow", "tracking_uri"))
mlflow.set_experiment(config.get("mlflow", "experiment_name"))
```

```python
for step in range(10):
    mlflow.log_metric("RMSE", float(2.0*step+1), step=step)
```

```python
with open("../../../model_checkpoints/all_epoch_metrics.json", "r") as f:
    metrics = json.load(f)
```
